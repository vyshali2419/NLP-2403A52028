```python
import spacy

nlp = spacy.load("en_core_web_sm")
print("Model loaded successfully!")
```

```python
sentences = [
    "Apple is looking at buying U.K. startup for $1 billion.",
    "Google is launching a new AI assistant this week.",
    "Tesla announced record profits last quarter."
]

docs = [nlp(sentence) for sentence in sentences]

for doc in docs:
    print(f"\nSentence: {doc.text}")
    for token in doc:
        print(token.text, token.pos_, token.dep_)
```

```python
import pandas as pd

data = []

for doc in docs:
    for token in doc:
        data.append([
            doc.text,
            token.text,
            token.lemma_,
            token.pos_,
            token.dep_,
            token.head.text
        ])

df = pd.DataFrame(data, columns=[
    "Sentence", "Token", "Lemma", "POS", "Dependency", "Head"
])

print(df.head(20))
```

```python
for doc in docs:
    print(f"\nSentence: {doc.text}")
    subject = [tok.text for tok in doc if tok.dep_ == "nsubj"]
    obj = [tok.text for tok in doc if tok.dep_ == "dobj"]
    root = [tok.text for tok in doc if tok.dep_ == "ROOT"]

    print("Subject:", subject)
    print("Verb (ROOT):", root)
    print("Object:", obj)
```

```python
from spacy import displacy

for doc in docs[:5]:
    displacy.render(doc, style="dep", jupyter=True)
```
