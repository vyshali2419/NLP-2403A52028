```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
```

```python
with open("rag_dataset.txt", "r") as f:
    documents = f.readlines()

documents = [doc.strip() for doc in documents]
print(documents[:5])
```

```python
embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
db = Chroma.from_texts(documents, embedding_model)
```

```python
retriever = db.as_retriever(search_kwargs={"k": 3})
```

```python
def simple_rag(query):
    docs = retriever.get_relevant_documents(query)
    answer = " ".join([d.page_content for d in docs])
    return answer
```

```python
query = "What is Artificial Intelligence?"

print("Answer:")
print(simple_rag(query))
```

```python
docs = retriever.get_relevant_documents(query)

print("\nRetrieved Documents:")
for d in docs:
    print("-", d.page_content)
```

```python
def ask_question(query):
    return simple_rag(query)

print(ask_question("Explain deep learning"))
```
