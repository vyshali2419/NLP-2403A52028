```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

# Create a custom summarizer class to mimic the pipeline behavior
class CustomSummarizer:
    def __init__(self, model, tokenizer):
        self.model = model
        self.tokenizer = tokenizer

    def __call__(self, text, max_length, min_length, do_sample):
        inputs = self.tokenizer([text], max_length=1024, return_tensors="pt", truncation=True)
        # Generate Summary
        summary_ids = self.model.generate(
            inputs["input_ids"],
            num_beams=4,
            max_length=max_length,
            min_length=min_length,
            early_stopping=True
        )
        summary_text = self.tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return [{'generated_text': summary_text}]

summarizer = CustomSummarizer(model, tokenizer)
print("Model Loaded Successfully!")
```

```python
texts = [
    """Artificial Intelligence is transforming industries by enabling machines
    to learn from data and perform tasks that traditionally required human intelligence.
    Applications include healthcare diagnosis, recommendation systems, and automation.
    However, ethical challenges such as bias and privacy remain concerns.""",

    """Climate change is one of the biggest global challenges. Rising temperatures,
    melting glaciers, and extreme weather events affect ecosystems and human life.
    Governments are promoting renewable energy and sustainability initiatives.""",

    """Python is a popular programming language known for readability and simplicity.
    It is widely used in web development, data science, machine learning, and automation.
    Its extensive libraries make development efficient.""",

    """Once upon a time, a young explorer traveled across oceans searching for
    hidden civilizations. Along the journey, challenges shaped his courage
    and wisdom, ultimately leading him to discover the value of friendship.""",

    """Hey, did you finish the assignment? Not yet, I was working on the coding part.
    The professor said we must also write analysis and compare outputs before submission."""
]
```

```python
results = []

for text in texts:
    summary = summarizer(text, max_length=60, min_length=20, do_sample=False)
    results.append(summary[0]['generated_text'])

# Display results
for i in range(len(texts)):
    print("\nOriginal Text:\n", texts[i])
    print("\nSummary:\n", results[i])
    print("-"*80)
```

```python
text_sample = texts[0]

short_summary = summarizer(text_sample, max_length=30, min_length=10, do_sample=False)
medium_summary = summarizer(text_sample, max_length=60, min_length=20, do_sample=False)
long_summary = summarizer(text_sample, max_length=100, min_length=40, do_sample=False)

print("Short Summary:\n", short_summary[0]['generated_text'])
print("\nMedium Summary:\n", medium_summary[0]['generated_text'])
print("\nLong Summary:\n", long_summary[0]['generated_text'])
```
