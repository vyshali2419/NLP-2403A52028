```python
# Install required libraries
!pip install transformers
!pip install torch

# Import libraries
from transformers import pipeline, AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
# English → French
model_name_fr = "Helsinki-NLP/opus-mt-en-fr"
tokenizer_fr = AutoTokenizer.from_pretrained(model_name_fr)
model_fr = AutoModelForSeq2SeqLM.from_pretrained(model_name_fr)

def translate_en_to_fr(text):
    inputs = tokenizer_fr(text, return_tensors="pt")
    outputs = model_fr.generate(**inputs)
    return tokenizer_fr.decode(outputs[0], skip_special_tokens=True)

# English → German
model_name_de = "Helsinki-NLP/opus-mt-en-de"
tokenizer_de = AutoTokenizer.from_pretrained(model_name_de)
model_de = AutoModelForSeq2SeqLM.from_pretrained(model_name_de)

def translate_en_to_de(text):
    inputs = tokenizer_de(text, return_tensors="pt")
    outputs = model_de.generate(**inputs)
    return tokenizer_de.decode(outputs[0], skip_special_tokens=True)
```

```python
sentences = [
    "I am learning Natural Language Processing.",
    "Machine learning is transforming the world.",
    "The exam will be conducted tomorrow.",
    "Can you help me with this assignment?",
    "Artificial Intelligence is the future of technology.",
    "The weather is very pleasant today.",
    "We are working on a translation project.",
    "This model converts English text into another language.",
    "Students must submit the assignment before deadline.",
    "Deep learning requires large amounts of data.",
    "How does a sequence to sequence model work?",
    "I completed my lab assignment successfully.",
    "The teacher explained the concept clearly.",
    "Programming becomes easy with regular practice.",
    "Natural language understanding is a difficult problem."
]
```

```python
print("ENGLISH TO FRENCH TRANSLATION\n")

for sentence in sentences:
    translated = translate_en_to_fr(sentence)
    print("English :", sentence)
    print("French  :", translated)
    print()
```

```python
print("ENGLISH TO GERMAN TRANSLATION\n")

for sentence in sentences:
    translated = translate_en_to_de(sentence)
    print("English :", sentence)
    print("German  :", translated)
    print()
```

```python
complex_sentences = [
    "If the model is trained on large data, it will produce better results.",
    "Although the problem is difficult, we will try to solve it.",
    "Can you explain how machine translation works in simple words?",
    "Students who complete the assignment early will get bonus marks.",
    "Artificial intelligence is growing rapidly, and it is used in many applications."
]

for sentence in complex_sentences:
    translated = translate_en_to_fr(sentence)
    print("English :", sentence)
    print("French  :", translated)
    print()
```
